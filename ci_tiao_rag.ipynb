```python
# data preprocessing
import pandas as pd
import numpy as np

# load xlsx file
df = pd.read_excel('citiao.xlsx')
df.head()
```

```python
# keep only point_name, item_id_cate, point_content, from_name
df = df[['point_name', 'item_id_cate', 'point_content', 'from_name']]
df.head()
```

```python
# remove html tags
from bs4 import BeautifulSoup
df['point_content'] = df['point_content'].apply(lambda x: BeautifulSoup(x, 'html.parser').get_text())
df.head()
```

```python
print(df['from_name'][0])
```

```python
# expand the from_name column to multiple columns, including from_level, from_title, from_date, from_docId, from_docSrc
# from_name is a string of dictionary, we need to convert it to dictionary first, they are in the format of [{"key":"value", "key":"value", ...}]
import ast

# Function to safely evaluate the string as a dictionary
def parse_dict_string(s):
    try:
        # Remove square brackets and evaluate the string as a dictionary
        return ast.literal_eval(s.strip('[]'))
    except:
        return {}

# Apply the parsing function to the 'from_name' column
df['from_name_dict'] = df['from_name'].apply(parse_dict_string)

# Extract the desired fields into new columns
def safe_get(x, key, default=''):
    if isinstance(x, dict):
        return x.get(key, default)
    elif isinstance(x, (tuple, list)) and len(x) > 0 and isinstance(x[0], dict):
        return x[0].get(key, default)
    else:
        return default

# Extract the desired fields into new columns
df['from_level'] = df['from_name_dict'].apply(lambda x: safe_get(x, 'level'))
df['from_title'] = df['from_name_dict'].apply(lambda x: safe_get(x, 'title'))
df['from_date'] = df['from_name_dict'].apply(lambda x: safe_get(x, 'publishDate'))
df['from_docId'] = df['from_name_dict'].apply(lambda x: safe_get(x, 'docId'))
df['from_docSrc'] = df['from_name_dict'].apply(lambda x: safe_get(x, 'docSrc'))

# Optionally, drop the intermediate 'from_name_dict' column
df = df.drop('from_name_dict', axis=1)
```

```python
# drop from_name column
df = df.drop('from_name', axis=1)
df.head()
```

# BM25 and Semantic Scores Combined

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from rank_bm25 import BM25Okapi
import numpy as np
import ollama
from sklearn.preprocessing import normalize

# Load your DataFrame (assuming it's named 'df')

# Step 1: Prepare documents
# documents = df.apply(lambda row: f"Term: {row['point_name']}\nCategory: {row['item_id_cate']}\nDefinition: {row['point_content']}\nLevel: {row['from_level']}\nTitle: {row['from_title']}\nDate: {row['from_date']}\nDocID: {row['from_docId']}\nSource: {row['from_docSrc']}", axis=1).tolist()
documents = df.apply(lambda row: f"术语: {row['point_name']}\n类别: {row['item_id_cate']}\n定义: {row['point_content']}\n级别: {row['from_level']}\n标题: {row['from_title']}\n日期: {row['from_date']}\n文档编号: {row['from_docId']}\n文档来源: {row['from_docSrc']}", axis=1).tolist()
# documents = df.apply(lambda row: f"术语: {row['point_name']}\n类别: {row['item_id_cate']}\n定义: {row['point_content']}\n级别: {row['from_level']}\n标题: {row['from_title']}", axis=1).tolist()

# Step 2: Create BM25 index
tokenized_documents = [doc.split() for doc in documents]
bm25 = BM25Okapi(tokenized_documents)

# Step 3: Load Sentence Transformer model
model = SentenceTransformer('shibing624/text2vec-base-chinese')

# Step 4: Compute embeddings for documents
doc_embeddings = model.encode(documents)
doc_embeddings = normalize(doc_embeddings)

# Step 5: Function to retrieve relevant documents
def retrieve_documents(query, top_k=3):
    # BM25 retrieval
    tokenized_query = query.split()
    bm25_scores = bm25.get_scores(tokenized_query)
    
    # Semantic similarity
    query_embedding = model.encode(query)
    semantic_scores = np.dot(doc_embeddings, query_embedding)
    
    combined_scores = 0.5 * bm25_scores + 0.5 * semantic_scores

    
    top_indices = combined_scores.argsort()[-top_k:][::-1]
    return [documents[i] for i in top_indices]

# Step 6: Function to generate answer using Ollama
def generate_answer(query, retrieved_docs):
    context = "\n\n".join(retrieved_docs)
    prompt = f"基于以下信息:\n\n{context}\n\n回答问题: {query}"
    
    response = ollama.generate(model='qwen2:7b', prompt=prompt)
    return response['response']

# Step 7: RAG pipeline
def rag_pipeline(query):
    retrieved_docs = retrieve_documents(query)
    answer = generate_answer(query, retrieved_docs)
    return answer, retrieved_docs

# Example usage
query = "采集器或计量点（电能表）所在的载波节点被称为是什么？"
answer, sources = rag_pipeline(query)

print(f"Question: {query}")
print(f"Answer: {answer}")
print("\nSource documents:")
for doc in sources:
    print(doc)
    print("---")
```

# FAISS

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from rank_bm25 import BM25Okapi
import numpy as np
import ollama
import faiss
from sklearn.preprocessing import normalize

# Load your DataFrame (assuming it's named 'df')

# Step 1: Prepare documents
# documents = df.apply(lambda row: f"Term: {row['point_name']}\nCategory: {row['item_id_cate']}\nDefinition: {row['point_content']}\nLevel: {row['from_level']}\nTitle: {row['from_title']}\nDate: {row['from_date']}\nDocID: {row['from_docId']}\nSource: {row['from_docSrc']}", axis=1).tolist()
documents = df.apply(lambda row: f"术语: {row['point_name']}\n类别: {row['item_id_cate']}\n定义: {row['point_content']}\n", axis=1).tolist()

# Step 2: Create BM25 index
tokenized_documents = [doc.split() for doc in documents]
bm25 = BM25Okapi(tokenized_documents)

# Step 3: Load Sentence Transformer model
model = SentenceTransformer('shibing624/text2vec-base-chinese')

# Step 4: Compute embeddings for documents
doc_embeddings = model.encode(documents)
doc_embeddings = normalize(doc_embeddings)

# Step 4: Index embeddings with FAISS
d = doc_embeddings.shape[1]  # Dimension of embeddings
index = faiss.IndexFlatIP(d)  # Use inner product (cosine similarity) for similarity search
index.add(doc_embeddings)  # Add document embeddings to the index

# Step 5: Function to retrieve relevant documents
def retrieve_documents(query, top_k=3):
    # Compute the query embedding
    query_embedding = model.encode([query])
    query_embedding = normalize(query_embedding)  # Normalize query embedding
    
    # Perform the search
    _, indices = index.search(query_embedding, top_k)
    
    return [documents[i] for i in indices[0]]

# Step 6: Function to generate answer using Ollama
def generate_answer(query, retrieved_docs):
    context = "\n\n".join(retrieved_docs)
    prompt = f"基于以下信息:\n\n{context}\n\n回答问题: {query}"
    
    response = ollama.generate(model='qwen2:7b', prompt=prompt)
    return response['response']

# Step 7: RAG pipeline
def rag_pipeline(query):
    retrieved_docs = retrieve_documents(query)
    answer = generate_answer(query, retrieved_docs)
    return answer, retrieved_docs

# Example usage
query = "什么是电容器？"
answer, sources = rag_pipeline(query)

print(f"Question: {query}")
print(f"Answer: {answer}")
print("\nSource documents:")
for doc in sources:
    print(doc)
    print("---")
```
